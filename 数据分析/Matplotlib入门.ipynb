```python
import matplotlib.pylab as plt
import random

# 绘制折线图
# 设置中文
plt.rcParams['font.sans-serif'] = ['SimHei']
x = [i for i in range(1,25)]
y = [random.randint(1,10) for i in x]
# 设置画布大小
plt.figure(figsize=(10,5))
plt.plot(x,y)

# 设置标题
plt.title('气温变化折线图',fontsize=20)
plt.xlabel('时间',fontsize=15)
plt.ylabel('气温',fontsize=15)
plt.xticks(x[::1])
y_yticks = [i for i in range(5,21)]
plt.yticks(y_yticks)
plt.show()
```
