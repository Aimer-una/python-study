```python

from matplotlib.axes import Axes
import matplotlib.pyplot as plt

# 展示中文
plt.rcParams['font.sans-serif'] = ['SimHei']

# 创建子图
figure,axes = plt.subplots(nrows=1,ncols=2,figsize=(20,6),dpi=100)

# 图一：柱状图
country = ['中国','美国','日本','印度','英国']
value = [1000,500,300,200,150]

axes1:Axes = axes[0]
axes1.bar(country,value,width=0.5,color='g')
axes1.set_title('世界石油储备',fontsize = 18)
axes1.set_xlabel('国家',fontsize = 16)
axes1.set_ylabel('油量（亿桶）',fontsize = 16)
```
