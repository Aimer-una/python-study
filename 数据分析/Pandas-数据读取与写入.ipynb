```python
import pandas as pd

# 读取数据
df = pd.read_csv('data/sales.csv',usecols=['订单号','产品类别','产品名称','销售数量','单价'])

# 数据处理
df['销售金额'] = df['销售数量'] * df['单价']
df

# 写入数据
df.to_csv('data/sales_result.csv',index=False)
```

```python
# 数据选择
df2 = pd.read_csv('data/sales.csv',usecols=['订单号','产品类别','产品名称','销售数量','单价'],index_col='订单号')
# 选择列
# 单列
df2['产品类别']

# 多列
df2[['产品类别','产品名称']]

# 选择行 iloc loc
# iloc 是根据行号来选择的（不包含结束位置）,语法:df.iloc[开始行号:结束行号,开始列号:结束列号]
df2.iloc[0:5:1]
df2.iloc[0:5]

# loc 是根据索引标签来选择的（包含结束位置）,语法:df.loc[开始行标签:结束行标签,开始列标签:结束列标签]
df2.loc[6805677496:6801664364:1]
```
