```python
# 数据选择
df2 = pd.read_csv('data/sales.csv',usecols=['订单号','产品类别','产品名称','销售数量','单价'],index_col='订单号')
# 选择列
# 单列
df2['产品类别']

# 多列
df2[['产品类别','产品名称']]

# 选择行 iloc loc
# iloc 是根据行号来选择的（不包含结束位置）,语法:df.iloc[开始行号:结束行号,开始列号:结束列号]
df2.iloc[0:5:1]
df2.iloc[0:5]

# loc 是根据索引标签来选择的（包含结束位置）,语法:df.loc[开始行标签:结束行标签,开始列标签:结束列标签]
df2.loc[6805677496:6801664364:1]
```

```python
df3 = pd.read_csv('data/sales.csv')
# 1.获取销售数量>=10的订单数据
df3[df3['销售数量']>=10]

# 2.获取产品类别为食品或者图书的订单数据
df3[df3['产品类别'].isin(['食品','图书'])]

# 3.获取单价在100-200之间的订单数据 -> 范围between -》默认包含边界(inclusive = "left")
df3[df3['单价'].between(100,200)]

# 4.获取销售数量>=8,并且单价>=100的订单数据
df3[(df3['销售数量']>=8) & (df3['单价']>=100)]

# 5.获取产品类别为食品或者服装，支付方式为微信或者支付宝
df3[(df3['产品类别'].isin(['食品','服装'])) & (df3['支付方式'].isin(['微信','支付宝']))]
```
