```python
import pandas as pd

df = pd.read_csv('data/sales.csv')
df.head(10)
# 查看缺失值
df.isnull()

# 删除缺失值
df.dropna() #删除缺失值所在行
# 删除缺失值所在列
df.dropna(axis=1)

# 填充缺失值
df.fillna("---")
df.ffill() #填充缺失值，填充上一行数据
df.bfill() #填充缺失值，填充下一行数据
```

```python
# 处理重复值
# 查看重复值
df.duplicated() # 查看重复值(所有列的数据都重复)

df.duplicated(subset=['订单号']) # 查看重复值(指定列的数据重复)

# 删除重复值
df.drop_duplicates(subset=['订单号']) #kepp = 'first' 保留第一行的重复值 keep = 'last' 保留最后一行重复值 keep = False 删除所有重复值
```

```python
# 异常值处理
# 查看异常值
df[df['单价'] < 0]

# 删除异常值
df.drop(df[df['单价'] < 0].index)

# 修复异常值
df['单价'] = df['单价'].abs() #绝对值
df
```

```python
# 数据格式处理
df['订单日期'] = df['订单日期'].str.replace('/', '-')
df
```
